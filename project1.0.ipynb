```python
columns = ['year_add','release_year','date_d','date_added']
df['year_add'] = pd.to_numeric(df['year_add'], errors='coerce')
df[columns].corr()
```

```python
# Create a new column to highlight 'Movie' type
df['highlight_type'] = df['type'].apply(lambda x: 'Movie' if x == 'Movie' else 'Other')

# Visualize the distribution of 'type' with 'Movie' highlighted
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='type', hue='highlight_type', palette={'Movie': 'red', 'Other': 'lightgray'})
plt.title('Distribution of Content Types with Movies Highlighted')
plt.xlabel('Content Type')
plt.ylabel('Count')
plt.show()
```

```python
sns.pairplot(data = df, diag_kind='kde',hue= 'type', palette ='Set1')
```

```python
genres = df['listed_in'].str.split(', ').explode()
genre_counts = genres.value_counts()
top_n = 10
top_genres = genre_counts.head(top_n)
explode = [0.1 if genre == 'Documentaries' else 0 for genre in top_genres.index]

plt.figure(figsize=(10, 10))
plt.pie(top_genres, labels=top_genres.index, autopct='%1.1f%%', startangle=90, explode=explode, pctdistance=0.85)
plt.title(f'Top {top_n} Content Categories on Netflix (Documentaries Highlighted)')
plt.axis('equal')
plt.show()
```
